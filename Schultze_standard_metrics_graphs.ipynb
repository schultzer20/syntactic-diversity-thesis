```python
def plot_classic_metrics(texts, bleu, ter, chrf):
    df = pd.DataFrame({
        "Text": texts,
        "BLEU": [bleu[t] for t in texts],
        "TER": [ter[t] for t in texts],
        "ChrF": [chrf[t] for t in texts],
    })

    plt.figure(figsize=(14,6))
    plt.plot(df["Text"], df["BLEU"], marker="o", label="BLEU")
    plt.plot(df["Text"], df["TER"], marker="o", label="TER")
    plt.plot(df["Text"], df["ChrF"], marker="o", label="ChrF")
    plt.xticks(rotation=45, ha="right")
    plt.ylim(0, 100)
    plt.ylabel("Score (0–100)")
    plt.title("BLEU, TER, and ChrF per Text")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_classic_metrics(texts, bleu_scores, ter_scores, chrf_scores)
```

```python
def plot_comet_metrics(texts, comet, xcomet_mt, xcomet_ht):
    df = pd.DataFrame({
        "Text": texts,
        "COMET": [comet[t] for t in texts],
        "XCOMET (MT)": [xcomet_mt[t] for t in texts],
        "XCOMET (HT)": [xcomet_ht[t] for t in texts],
    })

    plt.figure(figsize=(14,6))
    plt.plot(df["Text"], df["COMET"], marker="o", label="COMET")
    plt.plot(df["Text"], df["XCOMET (MT)"], marker="o", label="XCOMET (MT)")
    plt.plot(df["Text"], df["XCOMET (HT)"], marker="o", label="XCOMET (HT)")
    plt.xticks(rotation=45, ha="right")
    plt.ylim(0, 100)
    plt.ylabel("Score (0–100)")
    plt.title("COMET and XCOMET per Text")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_comet_metrics(texts, comet_scores, xcomet_mt_scores, xcomet_ht_scores)
```
